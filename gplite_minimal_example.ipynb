## Demo of _GPlite_  
### The Art of Gaussian Processes, Neurips 2021

```python
%load_ext autoreload
%autoreload 2
    
import numpy as np
import matplotlib.pyplot as plt
from gp_lite import *
```

```python
# instatiate
gp = gp_lite()
gp.init_hypers()
```

```python
# sample from the prior
i=300
gp.compute_posterior(dimension=i)
gp.sample(how_many=4)
gp.plot_samples()
```

```python
#produce data
N = 1000
x = 100*np.random.random(N) + 0.1*np.random.randn(N)
y = 20*np.cos(1.5*x)*np.cos(0.1*x) + 0.1*np.random.randn(N)

#subset
i = 120
x_sample = x[:i]
y_sample = y[:i]

#load and plot  data
gp.load(x_sample,y_sample)
gp.plot_data(v_axis_lims = 35)
```

```python
gp.compute_posterior(dimension=1000)
gp.plot_posterior(5,v_axis_lims = 35)
print(f'negative log-likelihood: {gp.nll()}')
```

```python
gp.train()
```

```python
gp.compute_posterior(dimension=1000)
gp.plot_posterior(5,v_axis_lims = 35)
print(f'negative log-likelihood: {gp.nll()}')
```
